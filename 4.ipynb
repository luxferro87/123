```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


# Вам нужно помочь секретарю автоматизировать работу.
# Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

def main():
    while True:
        user_input = input('\nВведите команду ')
        if user_input == 'p':
            print(people())
        elif user_input == 's':
            print(shelf())
        elif user_input == 'l':
            list_documents()
        elif user_input == 'as':
            add_new_shelf()
        elif user_input == 'ds':
            del_empty_shelf()
        elif user_input == 'ad':
            add_doc()

        elif user_input == 'd':
            delete()

        elif user_input == 'm':
            print(move())
        elif user_input == 'q':
            print('До свидания')
            break


def people():
    """
    p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
    """

    number = input('Введите номер документа владельца ')
    for doc in documents:
        if doc['number'] == number:
            owner = doc['name']
            return f'Владелец документа: {owner}'
    return f'Документ не найден в базе'


def shelf():
    """
    s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
    """

    number = input('Введите номер документа ')
    for shelf_num, docs in directories.items():
        if number in docs:
            return f'Документ хранится на полке: {shelf_num}'
    return 'Документ не найден в базе'


def list_documents():
    """
    l– list – команда, которая выведет список всех документов в формате passport "2207 876234" "Василий Гупкин";
    """
    for row in documents:
        doc_number, doc_type, doc_owner = row['number'], row['type'], row['name']
        print(f' №:{doc_number}, тип: {doc_type}, владелец: {doc_owner}')


def add_new_shelf():
    """
    as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
    """
    user_input = input('Введите, пожалуйста, номер новой полки ')
    if directories.get(user_input) is None:
        directories.setdefault(user_input, [])
        print('Полка добавлена в текущий перечень полок:', ', '.join(list(directories.keys())))
    else:
        print('Такая полка уже существует')


def del_empty_shelf():
    """
    Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)

    """
    shelf_number = input('Укажите номер пустой полки ')
    shelfs = ", ".join(list(directories.keys()))

    if shelf_number not in directories.keys():
        print(f'Такой полки не существует. Текущий перечень полок {shelfs}')
    elif shelf_number in directories.keys() and len(directories[shelf_number]) > 0:
        print(f'Не могу удалить эту полку, на ней лежат документы. Текущий перечень полок: {shelfs}')
    else:
        del directories[shelf_number]
        print(f'Полка №{shelf_number} удалена из каталога')


def add_doc():
    """
    Пользователь по команде “ad” может добавить новый документ в данные

    """
    new_type = input('Укажите тип документа ')
    new_num = input('Введите, пожалуйста, новый документ ')
    new_owner = input('Укажите владельца документа ')
    new_shelf = input('Введите новую полку для хранения ')

    new_document = {'type': new_type, 'number': new_num, 'name': new_owner}
    if new_document in documents:
        print(f'Такой документ уже заведен в базе')
    else:
        documents.append(new_document)
        if new_shelf not in directories.keys():
            print('Такой полки не существует. Добавьте полку командой as.')
            print('Текущий список документов:')


def delete():
    """
      d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок.
      Предусматривает сценарий, когда пользователь вводит несуществующий документ;
    """

    user_input = input('Введите номер документа, который необходимо удалить из каталога и перечня полок ')
    in_documents = False
    for doc in documents:

        if doc['number'] == user_input:
            doc_num = doc['number']
            in_documents = True
            # Удаляем указанный документ из списка документов
            documents.remove(doc)
            print('Строка c документом {} успешна удалена'.format(doc_num))
            # Теперь мы удаляем указанный документ с полки, на которой он находится
            for shelf, content in directories.items():
                if doc_num in content:
                    content.remove(doc_num)
    if not in_documents:
            print('Указанный документ отсутствует в базе')


def move():
    """

    Пользователь по команде “m” может переместить документ с полки на полку

    """
    # Проверка есть ли указанный документ среди текущих документов
    input_user = input('Введите, пожалуйста, номер документа: ')
    doc_for_move = None # Заводим переменную для документа, который попытаемся переместить

    for doc in documents:
        if doc['number'] == input_user:
            doc_for_move = doc['number'] # Если док-т есть, записываем его в переменную
            break

    if doc_for_move is None:
        return f'Документ не найден в базе.' # В случае, если док-т не нашли выводим соответствующее сообщение.
    else:
        # Найденный документ перемещаем на указанную полку
        new_shelf = input('Введите новый номер полки ')
        if new_shelf not in directories.keys():
            shelfs = ', '.join(list(directories.keys()))
            return f'Такой полки не существует. Текущий перечень полок:{shelfs}' # По ключу вытаскиваем значение с номером полки
        else:
            for shelf, content in directories.items():
                if doc_for_move in content:
                    # Если мы нашли полку с указанным документом, то мы на этой полке удаляем этот документ
                    content.remove(doc_for_move)
                    # Переносим документ на новую полку
                    goal_shelf = directories[new_shelf] # Список с документами
                    goal_shelf.append(doc_for_move)
                    return f'Документ: {doc_for_move}, перемещен на полку {new_shelf}'

main()
```
